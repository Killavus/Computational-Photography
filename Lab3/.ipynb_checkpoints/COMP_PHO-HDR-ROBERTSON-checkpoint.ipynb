```python
import numpy as np
import imageio
```

```python
imageFiles = [
    './HDRSeq/_MG_4460.tiff',
    './HDRSeq/_MG_4461.tiff',
    './HDRSeq/_MG_4462.tiff'
]

exposureTime = [0.125, 0.033333, 0.5]

images = map(lambda path: imageio.imread(path) / 65535.0, imageFiles)
```

```python
for i in range(len(images)):
    subplot(1, len(images), i+1)
    imshow(images[i])
```

```python
xPoints = np.linspace(0.0, 1.0, 1000)
yPoints = np.exp(-4 * (np.power(xPoints - 0.5, 2) / 0.25))

plot(xPoints, yPoints)
```

```python
def robertsonWeights(images):
    result = []
    for i in range(len(images)):
        weightsMatrix = np.exp(-4 * (np.power(images[i] - 0.5, 2) / (0.5 ** 2)))
        result.append(weightsMatrix)
    return result

weights = robertsonWeights(images)
```

```python
def robertsonHDR(images, weights, exposureTime):
    hdrResultTop = np.zeros(shape=images[0].shape)
    hdrResultBottom = np.zeros(shape=images[0].shape)
    for index in range(len(images)):
        hdrResultTop += (images[index] * weights[index] * exposureTime[index]) 
        hdrResultBottom += (weights[index] * (exposureTime[index] ** 2))
        
    return hdrResultTop / hdrResultBottom
```

```python
hdrImage = robertsonHDR(images, weights, exposureTime)
```

```python
def saveHDR(filename, image):
    f = open(filename, "wb")
    f.write("#?RADIANCE\n# Made with Python & Numpy\nFORMAT=32-bit_rle_rgbe\n\n")
    f.write("-Y {0} +X {1}\n".format(image.shape[0], image.shape[1]))
    
    brightest = np.maximum(np.maximum(image[...,0], image[...,1]), image[...,2])
    mantissa = np.zeros_like(brightest)
    exponent = np.zeros_like(brightest)
    np.frexp(brightest, mantissa, exponent)
    scaled_mantissa = mantissa * 256.0 / brightest
    rgbe = np.zeros((image.shape[0], image.shape[1], 4), dtype=np.uint8)
    rgbe[...,0:3] = np.around(image[...,0:3] * scaled_mantissa[...,None])
    rgbe[...,3] = np.around(exponent + 128)
    
    rgbe.flatten().tofile(f)
    f.close()

saveHDR('sec1.hdr', hdrImage)
```

```python
jpegPaths = [
    './HDRSeq/_MG_4460.JPG',
    './HDRSeq/_MG_4461.JPG',
    './HDRSeq/_MG_4462.JPG'
]

jpegs = map(lambda path: imageio.imread(path) / 255.0, jpegPaths)

def applyGamma(image, gamma = 1/2.2):
    red = image[:, :, 0]
    green = image[:, :, 1]
    blue = image[:, :, 2]

    red = np.power(red, gamma)
    green = np.power(green, gamma)
    blue = np.power(blue, gamma)

    result = np.empty(shape=(image.shape[0], image.shape[1], 3))
    result[:, :, 0] = red
    result[:, :, 1] = green
    result[:, :, 2] = blue
    
    return result

jpegs = map(lambda jpeg: applyGamma(jpeg, 2.2), jpegs)
jpegWeights = robertsonWeights(jpegs)

jpegHDRImage = robertsonHDR(jpegs, jpegWeights, exposureTime)
saveHDR('sec1-jpeg.hdr', jpegHDRImage)
```
